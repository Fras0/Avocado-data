```python
#importing some needed libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
da  = pd.read_csv('D:/Data/avocado.csv')
```

```python
da.head()
```

```python
da = da.drop(columns = ['Unnamed: 0']) #removing unnamed column
da['year'] = da['year'].astype('str') 
```

```python
da['year'].unique()
```

```python
da.info()
```

```python
sns.distplot( da['AveragePrice'])
```

```python
plt.subplot(2,2,1)
plt.title("average price for avocato in 2015")
plt.hist(da.loc[da['year'] == '2015','AveragePrice'])

plt.subplot(2,2,2)
plt.title("average price for avocato in 2016")
plt.hist(da.loc[da['year'] == '2016','AveragePrice'])

plt.subplot(2,2,3)
plt.title("average price for avocato in 2017")
plt.hist(da.loc[da['year'] == '2017','AveragePrice'])

plt.subplot(2,2,4)
plt.title("average price for avocato in 2018")
plt.hist(da.loc[da['year'] == '2018','AveragePrice'])

plt.tight_layout()
plt.show()
```

```python
plt.bar(da['year'],da['Total Volume'])
plt.title('Avocado sales over the years')
```

```python
plt.subplot(2,2,2)
plt.bar(da['year'],da['4046'])
plt.title('PLU 4046 sales')

plt.subplot(2,2,4)
plt.bar(da['year'],da['4225'])
plt.title('PLU 4225 sales')


plt.subplot(2,2,3)
plt.bar(da['year'],da['4770'])
plt.title('PLU 4770 sales')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(da['4046'],da['AveragePrice'],marker = 'o',color = 'red',label = 'PLU 4046')
plt.scatter(da['4225'],da['AveragePrice'],marker = 'o',color = 'green',label = 'PLU 4225')
plt.scatter(da['4770'],da['AveragePrice'],marker = 'o',color = 'blue',label = 'PLU 4770')
plt.legend(loc = 'upper right')
plt.title('PLU prices')
plt.xlabel('avocado sold')
plt.ylabel('Average Price')
plt.show()
```

```python
sns.FacetGrid(da,row='type',col ='year').map(plt.scatter,'AveragePrice','Total Volume').add_legend()
```

```python
plt.figure(figsize=(15, 7))
plt.xticks(rotation=90)
sns.boxplot(x = da['region'], y = da['AveragePrice'],data = da) 
```
